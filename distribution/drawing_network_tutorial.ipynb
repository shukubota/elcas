```python
# 以下のファイルからグラフを描画するのに必要な関数をimportします
# ファイル内でnumpyなどはインストールされているので宣言は不要です
from src.networkx_func import *
```

```python
# 簡単なネットワークを隣接リスト形式で記述します．
network_list = [[1, 2, 3, 4, 5, 6, 7, 8], [0, 2, 3, 4, 5, 9], [1, 3, 4, 6], [0, 1, 2], [0, 1, 2, 5, 6, 7, 8, 9], [0, 1, 4, 9], [0, 2, 7], [0, 4, 6, 8], [0, 4, 7], [1, 4, 5]]

# ノードの大きさを調整する関数の例
# 関数も自分で作ることができます
def log_700(x):
    return np.log(x)*700
def mul_700(x):
    return x*700
```

```python
# とりあえずグラフを描画してみます．
# OutputNGraph()という関数を用意しました．
# 引数は隣接リスト，表示するノード数,重み関数,ノード間の反発力(0~1までの少数),書き出すファイル名(今回は書き出さないので'NULL')
OutputNGraph(network_list, len(network_list), mul_700, 0.7, 'NULL')
```

```python
# 関数の引数を自分で変えてみて遊んでみましょう
# 簡単やん！という人はnetworkx_func.py内の関数を自分で弄ってグラフのレイアウトを変更してみましょう
OutputNGraph(network_list, len(network_list)-1, mul_700, 2.3, 'NULL')
```

```python
# 先程のOutputNGraph()は成長を始めてからノードがK個になった時点でのグラフを表示していました
# ではノードを意図的に消して図にしたい時の関数を考えましょう
```

```python
#　次の関数は次数が小さいものは消して表示するものです
# 元の比較してみると確かに次数が小さいノードが削除されていることがわかります
OutputNGraphDescend(network_list, len(network_list) - 2, mul_700, 0.7, 'NULL')
```

```python
#　次の関数は次数が大きいものは消して表示するものです
# 元の比較してみると確かに次数が大きいノードが削除されていることがわかります
OutputNGraphAscend(network_list, len(network_list) - 2, mul_700, 0.7, 'NULL')
```

```python
# またこの2つのグラフで少し遊んでみましょう
OutputNGraphDescend(network_list, len(network_list) - 2, mul_700, 0.7, 'NULL')
```

```python
# ネットワークを作成した隣接リストを読み込んで関数を定義，グラフを表示し保存する流れを最後にお見せして終わりたいと思います
# データの読み込みに時間がかかるので読み込みとグラフ描画は別のセルのほうがいいかもしれません

# sample1
nwdata1 = np.load('src/nwdatam1.npy')

def func1(x):
    return (x**1.5)*5.0

OutputNGraph(nwdata1, 1000, func1, 0.03, 'm1BAmodel.png')
```
